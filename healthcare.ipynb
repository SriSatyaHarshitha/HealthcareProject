```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

# Assuming df_h is your dataframe and 'Showed_up' is the target variable
# Step 1: Check if there are non-numeric columns
print(df_h.select_dtypes(include=['object']).columns)  # Print all non-numeric columns

# Step 2: Drop or encode non-numeric columns (e.g., 'PatientId', 'AppointmentID', 'Neighbourhood', etc.)
# In case 'PatientId' and 'AppointmentID' are identifiers and not relevant for the model, drop them.
df_encoded = df_h.drop(columns=['PatientId', 'AppointmentID'])  # Drop irrelevant columns (or use your own logic)

# Step 3: One-Hot Encoding for categorical columns
df_encoded = pd.get_dummies(df_encoded, drop_first=True)  # Automatically converts all categorical columns

# Step 4: Ensure there are no missing values
df_encoded.fillna(df_encoded.mean(), inplace=True)  # Fill missing values with the mean (for numerical columns)

# Step 5: Select features (X) and target (y)
X = df_encoded.drop(columns=['Showed_up'])  # Drop the target column
y = df_encoded['Showed_up']  # Target variable

# Step 6: Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Step 7: Initialize and train the Logistic Regression model
model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)

# Step 8: Make predictions
y_pred = model.predict(X_test)

# Step 9: Evaluate the model
print(f'Accuracy: {accuracy_score(y_test, y_pred)}')
print(f'Confusion Matrix:\n{confusion_matrix(y_test, y_pred)}')
print(f'Classification Report:\n{classification_report(y_test, y_pred)}')
```

```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_curve, auc, confusion_matrix

# Function to compute evaluation metrics
def evaluate_model(model_name, y_test, y_pred, y_proba=None):
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred, pos_label=1)
    recall = recall_score(y_test, y_pred, pos_label=1)
    f1 = f1_score(y_test, y_pred, pos_label=1)
    
    print(f"\n{model_name} Evaluation:")
    print(f"Accuracy: {accuracy:.4f}")
    print(f"Precision: {precision:.4f}")
    print(f"Recall: {recall:.4f}")
    print(f"F1-Score: {f1:.4f}")
    print(f"Confusion Matrix:\n{confusion_matrix(y_test, y_pred)}")
    
    if y_proba is not None:
        fpr, tpr, _ = roc_curve(y_test, y_proba)
        roc_auc = auc(fpr, tpr)
    else:
        fpr, tpr, roc_auc = None, None, None
    
    return {
        "Model": model_name,
        "Accuracy": accuracy,
        "Precision": precision,
        "Recall": recall,
        "F1-Score": f1,
        "FPR": fpr,
        "TPR": tpr,
        "AUC": roc_auc
    }

# Evaluate all models
results = []

# Logistic Regression
y_pred_lr = model.predict(X_test)  # Replace 'model' with your logistic regression model variable
y_proba_lr = model.predict_proba(X_test)[:, 1]
results.append(evaluate_model("Logistic Regression", y_test, y_pred_lr, y_proba_lr))

# Random Forest
y_pred_rf = rf_model.predict(X_test)
y_proba_rf = rf_model.predict_proba(X_test)[:, 1]
results.append(evaluate_model("Random Forest", y_test, y_pred_rf, y_proba_rf))

# SVM
y_pred_svm = model.predict(X_test)  # Replace 'model' with your SVM model variable
if hasattr(model, "predict_proba"):
    y_proba_svm = model.predict_proba(X_test)[:, 1]
else:
    y_proba_svm = None
results.append(evaluate_model("SVM", y_test, y_pred_svm, y_proba_svm))

# Convert results to a DataFrame for graphical representation
results_df = pd.DataFrame(results)

# Plotting Bar Chart for Accuracy, Precision, Recall, and F1-Score
metrics = ["Accuracy", "Precision", "Recall", "F1-Score"]
plt.figure(figsize=(10, 6))
for i, metric in enumerate(metrics):
    plt.bar(
        results_df["Model"] + f" ({metric})",
        results_df[metric],
        label=metric,
        alpha=0.7
    )

plt.title("Model Performance Comparison")
plt.ylabel("Scores")
plt.xticks(rotation=45, ha='right')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

# Assuming X and y are defined and preprocessed correctly, as in your previous code

# Step 1: Create a function to get user input for each feature
def get_user_input():
    print("Please enter the following details:")
    
    gender = input("Gender (M/F): ").strip().lower()  # For categorical input (object type)
    scheduled_day = int(input("Scheduled Day (as integer, e.g., 15 for 15th): "))
    appointment_day = int(input("Appointment Day (as integer, e.g., 20 for 20th): "))
    age = int(input("Age (as integer, e.g., 30): "))
    neighbourhood = input("Neighbourhood (e.g., 'Downtown'): ").strip()
    scholarship = bool(input("Scholarship (True/False): "))
    hypertension = bool(input("Hypertension (True/False): "))
    diabetes = bool(input("Diabetes (True/False): "))
    alcoholism = bool(input("Alcoholism (True/False): "))
    handicap = bool(input("Handicap (True/False): "))
    sms_received = bool(input("SMS received (True/False): "))
    date_diff = int(input("Date difference (as integer, e.g., 5): "))
    age_group = input("Age group (e.g., 'Young', 'Middle-aged', 'Old'): ").strip()
    day_of_week = input("Day of the week (e.g., 'Monday', 'Tuesday'): ").strip()
    health_risk = bool(input("Health Risk (True/False): "))
    age_health_risk_interaction = int(input("Age * Health Risk Interaction (as integer, e.g., 15): "))

    # Create a dictionary of the inputs
    user_input = {
        'Gender': gender,
        'ScheduledDay': scheduled_day,
        'AppointmentDay': appointment_day,
        'Age': age,
        'Neighbourhood': neighbourhood,
        'Scholarship': scholarship,
        'Hypertension': hypertension,
        'Diabetes': diabetes,
        'Alcoholism': alcoholism,
        'Handicap': handicap,
        'SMS_received': sms_received,
        'Date.diff': date_diff,
        'Age_group': age_group,
        'age_group': age_group,
        'day_of_week': day_of_week,
        'Health_Risk': health_risk,
        'Age_HealthRisk_interaction': age_health_risk_interaction
    }
    
    return user_input

# Step 2: Preprocessing Input Data
def preprocess_input(user_input):
    # Convert the inputs into a format suitable for the model (e.g., encoding categorical variables)
    
    # For simplicity, you can use pandas' `get_dummies` for categorical features like 'Gender', 'Neighbourhood', etc.
    user_df = pd.DataFrame([user_input])
    
    # Example: Encoding categorical features, assume necessary encoding steps here.
    # E.g., Gender and Neighbourhood might need one-hot encoding.
    user_df = pd.get_dummies(user_df, drop_first=True)
    
    # Ensure the input has the same feature columns as your training data
    user_df = user_df.reindex(columns=X.columns, fill_value=0)  # Match the column order in your training data
    
    return user_df

# Step 3: Make Prediction Using the Chosen Model
def make_prediction(model, user_input):
    user_df = preprocess_input(user_input)
    
    # Predict using the trained model
    prediction = model.predict(user_df)
    
    # Return prediction result
    return prediction

# Assuming you have already trained the models like Logistic Regression, RandomForest, and SVC
# Example usage with Logistic Regression model:
# Replace `model` with the specific model you're using (e.g., LogisticRegression, RandomForestClassifier, or SVC)
# The models are assumed to be already trained, as shown in your provided code.

# Example: Logistic Regression Model
model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)  # Train your model using your training data

# Get user input
user_input = get_user_input()

# Make prediction
prediction = make_prediction(model, user_input)

# Output prediction result
print(f"Prediction result: {prediction[0]}")
```
